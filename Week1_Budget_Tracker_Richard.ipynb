## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income
- Takes up to 3 expenses
- Calculates what's left over
- Warns the user if the budget goes negative

> 💡 **Pro tip:** The more specific your prompt, the less clean-up you'll need to do afterwards. "Write a budget tracker" gets you something generic; describing the exact inputs, outputs, and edge cases gets you something closer to what you actually need.


```python
# Write your detailed prompt here:
"""
My Budget Tracker Prompt:
rite a Python program for a basic budget tracker that:
1. Prompts the user to enter their monthly income (as a number).
2. Prompts the user to enter up to 3 separate expense amounts.
3. Calculates the total expenses and the remaining balance (Income - Total Expenses).
4. Prints a summary showing income, total expenses, and remaining budget.
5. Displays a warning message if the remaining balance goes below zero (budget goes negative).
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]
def main():
    print("--- SIMPLE BUDGET TRACKER ---\n")

    # 1. Get user's monthly income
    try:
        income = float(input("Enter your monthly income: $"))
    except ValueError:
        print("Invalid input. Setting income to $0.00")
        income = 0.0

    # 2. Get up to 3 expenses
    print("\nEnter your expenses (up to 3):")
    total_expenses = 0.0

    for i in range(1, 4):
        user_input = input(f"Enter expense #{i} (or press Enter to skip): $")

        # If user presses enter without typing a number, break early
        if user_input.strip() == "":
            break

        try:
            expense = float(user_input)
            total_expenses += expense
        except ValueError:
            print(f"  Invalid amount for expense #{i}, skipped.")

    # 3. Calculate remaining balance
    remaining_balance = income - total_expenses

    # 4. Print summary output
    print("\n" + "=" * 35)
    print("          BUDGET SUMMARY          ")
    print("=" * 35)
    print(f"Monthly Income:    ${income:,.2f}")
    print(f"Total Expenses:    ${total_expenses:,.2f}")
    print("-" * 35)
    print(f"Remaining Balance: ${remaining_balance:,.2f}")
    print("=" * 35)

    # 5. Warning check if budget is negative
    if remaining_balance < 0:
        print("\n⚠️ WARNING: Your budget is in the negative!")
        print(f"   You are overspending by ${abs(remaining_balance):,.2f}.")
    else:
        print("\n Great job! You are within your budget.")

if __name__ == "__main__":
    main()
```

## Part 4: Enhancement Challenge

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 A detailed budget summary with advice


```python
# Document your enhancement choice and the prompt you used
enhancement_choice = "I chose to add: [Percentage of income spent]"
enhancement_prompt = """
Modify the budget tracker script to calculate and display what percentage of the total monthly income is taken up by expenses.
Additionally, add a budget health assessment based on the spending ratio:
- Less than 50%: Excellent budget management
- 50% to 80%: Good balance
- 81% to 99%: High spending zone
- 100% or more: Over budget alert
"""

print(f"🚀 Enhancement: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
# income = float(input("Enter your monthly income: $"))

income_input = input("Enter your monthly income: $").strip().replace("$", "").replace(",", "")
income = float(income_input) if income_input else 0.0

expenses = []
for i in range(1, 4):
    val = input(f"Enter expense #{i} (or press Enter to skip): $").strip().replace("$", "").replace(",", "")
    if not val:
        break
    expenses.append(float(val))

total_expenses = sum(expenses)
remaining = income - total_expenses
spending_ratio = (total_expenses / income * 100) if income > 0 else 0.0

print("\n--- BUDGET SUMMARY ---")
print(f"Income:           ${income:,.2f}")
print(f"Total Expenses:   ${total_expenses:,.2f}")
print(f"Spending Ratio:   {spending_ratio:.1f}%")
print(f"Remaining:        ${remaining:,.2f}")

print("\n--- HEALTH ASSESSMENT ---")
if remaining < 0:
    print(f"🚨 OVER BUDGET ALERT! You are over by ${abs(remaining):,.2f}.")
elif spending_ratio >= 81:
    print("⚠️ HIGH SPENDING ZONE! Consider cutting down non-essential costs.")
elif spending_ratio >= 50:
    print("👍 GOOD BALANCE! Moderate spending habits.")
else:
    print("🌟 EXCELLENT BUDGET MANAGEMENT! Great savings potential.")
```
